A linear regression model makes a prediction by simply computing a weighted sum of the input features and 

a constant called the bias term(Also known as the intercept term).

Below is the form for multiple linear regression.

$\hat y = \theta_0+\theta_1x_1+\theta_2x_2+...+\theta_nx_n$

where

$\hat y$  is the prdicted value

$n$ is the number of features

$x_i$ is the $i^{th}$ feature

$\theta_j$ nis the $j^{th}$ model parameter (including  the bias term $\theta_0$ and the feature weights $\theta_1,\theta_2,...,\theta_n$)

...............................................................................................................................



Linear Regression model in vectorized form

$\hat y = h_\theta(x) = \theta^T.x$

* $\theta$ is the model's parameter vector,conaining the bis term $\theta_0$, and the feature weights $\theta_1$
 to $\theta_n$


* $\theta^T$ is the transpose of $\theta$

* $x$ is the instance's feature vector, containing $x_0$ to $x_n$,with $x_0$ always equal to 1.

* $\theta^T.x$  is the dot product of $\theta^T$ and x.

* $h_\theta$ is the hypothesis function,using model parameters $\theta$

...............................................................................................................................

To train a Linear Regression Model, one needs to find a value of  $\theta$ that minimizes MSE.

Most Common Performance measure of a Linear regression model is MSE(Mean Squared Error)

The MSE of a linear regression hypothesis $ h_\theta$ on a training set $X$ is given by

MSE(X,$h_\theta$) =$\frac{1}{m}\sum_{i=1}^m (\theta^T.x^{(i)}-y^{(i)})^2$

MSE cost function for a linear regression is a convex function. That implies MSE has no local minima ,just one global minimum.


...............................................................................................................................





The Normal Equation

To find a value of  $\theta$ that minimizes MSE (cost function),there is a closed form solution which gives the result directly.

This is called the Normal Equation.

$\hat \theta = (X^T.X)^{-1}.X^T.y$ 


* $\hat \theta$ is the value of $\theta$ that minimizes the cost function

* $x$ is the instance's feature vector, containing $x_0$ to $x_n$,with $x_0$ always equal to 1.

* $y$ is the vector of target values conatining $y^{(1)}$ to $y^{(m)}$








```python


## Generating a random linear data

import numpy as np
import matplotlib.pyplot as plt
X=2* np.random.rand(100,1)
y= 4+3*X+ np.random.randn(100,1)
plt.scatter(X,y)
```

```python
## Implementation of Linear Regression using Normal Equation
X_b=np.c_[np.ones((100,1)),X] # adding X0=1 to each instance
theta_best=np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
theta_best
```

Gradient Descent

Gradient descent is a very generic optimization algorithm capable of finding optimal solutions.

The general idea of gradient Descent is to tweak parameters iteratively in order to minimize a cost function(example MSE).




An important parameter in Gradient Descent is the size of steps,determinced by the learning parameter.

If learning rate is too small,then the algorithm will have to go through many iterations to converge,which will take a long time.

On the other hand if learning rate is too high, would make the algorithm diverge,with larger and larger values,failing to finf a good solution.



Batch Gradient Descent

To compute Gradient descent one needs to compute the gradient of the cost function (MSE) with regards to each
model parameter $\theta_j$

Below is the partial dervivative of the cost function


$\frac{\partial MSE(\theta)}{\partial\theta_j}$ =$\frac{2}{m}\sum_{i=1}^m (\theta^T.x^{(i)}-y^{(i)})x_j^{(i)}$


Gradient Vector form of the cost function

$\nabla  MSE(\theta)$ =$\frac{2}{m}$$X^T.(X.\theta-y)$

In Batch gradient descent we consider the entire training data at every step.

Hence it is slow at large training sets.




Convergence Theorem:

This theorem helps us compute the next Gradient descent step.

It is given by the formula.



$\theta^{(next step)}=\theta-\eta\nabla_{\theta} MSE(\theta)$

Here 

$\eta$ is the learning rate





```python
eta=0.1
n_iterations=1000
m=100
theta=np.random.randn(2,1)

for iteration in range(n_iterations):
    gradients= 2/m*X_b.T.dot(X_b.dot(theta)-y)
    theta=theta-eta*gradients
    
```
